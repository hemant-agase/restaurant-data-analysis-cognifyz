## Task 1: Restaurant Reviews

- Analyze the text reviews to identify the most common positive and negative keywords.

- Calculate the average length of reviews and explore if there is a relationship between review length and rating.


```python
# 1. identify the most unique positive and negative keywords.
unique_words = set(df["Rating text"])
unique_words
```

```python
import nltk
import string
from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
# download resources
nltk.download('vader_lexicon')
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')

# 1. text analyzer 

sia= SentimentIntensityAnalyzer()
stop_words= set(stopwords.words('english'))
punch_remove= str.maketrans('','', string.punctuation)

# pos and neg words 
pos, neg = [],[]

for text in df['Rating text'].astype(str):
    tokens= [w for w in word_tokenize(text.lower().translate(punch_remove)) if w not in stop_words] 
    score = sia.polarity_score(text)['compound']
    (pos if score>= 0.05 else neg).extend(tokens)

# freqency count
pos_wc = pd.Series(pos).value_counts().head(10)
neg_wc = pd.Series(neg).value_counts().head(10)

print("common positive keywords \n",pos_wc)
print("common negative keyword \n",neg_wc)
```

```python
import nltk
import string
from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.tokenize import word_tokeniz
from nltk.corpus import stopwords

# Download required resources
nltk.download('vader_lexicon')
nltk.download('punkt')
nltk.download('stopwords')

# Init analyzer and stopwords
sia = SentimentIntensityAnalyzer()
stop_words = set(stopwords.words('english'))
punct_remover = str.maketrans('', '', string.punctuation)

# Preprocess + classify words
pos, neg = [], []

for text in df['Rating text'].astype(str):
    tokens = [w for w in word_tokenize(text.lower().translate(punct_remover)) if w not in stop_words]
    score = sia.polarity_scores(text)['compound']
    (pos if score >= 0.05 else neg).extend(tokens)

# Frequency count
pos_wc = pd.Series(pos).value_counts().head(10)
neg_wc = pd.Series(neg).value_counts().head(10)

# Output
print(" Most common positive keywords:\n", pos_wc)
print("\n Most common negative keywords:\n", neg_wc)
```

```python
# show in chart 
fig, ax = plt.subplots(1,2, figsize=(10,6))
plt.suptitle(' Positive and Negative Keywords')
# positive chart 
ax[0].pie(pos_wc.head(10), labels=pos_wc.head(10).index, autopct= '%1.1f%%')
ax[0].set_title("Most Common Positive Keywords")

# Negative chart 
ax[1].pie(neg_wc.head(10), labels=neg_wc.head(10).index, autopct= '%1.1f%%')
ax[1].set_title("Most Common Negative Keywords")

plt.tight_layout()
plt.show()
```

## Task 2

### Task: Votes Analysis
- Identify the restaurants with the highest and lowest number of votes.
- Analyze if there is a correlation between the number of votes and the rating of a restaurant.

```python
# 1. clean data
df_clean= df.dropna(subset= ['Votes'])

# 2. find highest votes Restaurants
max_vote_val= df_clean['Votes'].max()
max_vote= df_clean[df_clean["Votes"] == max_vote_val]
highest_votes = max_vote[["Restaurant Name","Votes"]]
print("Highest votes with Restaurants \n",highest_votes)

# 3. find lowest votes Restaurants
min_vote_val = df_clean['Votes'].min()
min_vote = df_clean[df_clean['Votes'] == min_vote_val]
lowest_votes = min_vote[['Restaurant Name','Votes']]
print("\n lowest votes with Restaurants \n", lowest_votes.head(10))

# 4.There are too many restaurants with the lowest number of votes
count_min_votes =len(min_vote)
print("\n Total Restaurants with lowest vote", min_vote_val ,":",count_min_votes)
```

### Conclusion:

- Since 1,094 restaurants have zero votes, the dataset is heavily skewed toward low engagement.
- This makes it unreliable to directly analyze correlation between votes and ratings across the full dataset.
- Therefore, we cannot accurately determine any meaningful correlation between votes and rating unless we remove or separately handle these zero-vote entries.


### if we remove 0 votes then correlate 

```python
# 1. remove zero votes restaurants 
df_filter = df_clean[(df_clean["Votes"]>0) & (df_clean["Aggregate rating"]>0)]

# 2. calculate correlation 
from scipy.stats import pearsonr
correlation, _ = pearsonr(df_filter["Votes"],df_filter["Aggregate rating"])
print(f'Correlation between rating and votes: {correlation}')
print("\n A correlation value of 0.41 indicates a moderate positive relationship between votes and ratings.")

# visualization 
plt.figure(figsize=(8, 6))
sns.regplot(x='Votes', y='Aggregate rating', data=df_filter, scatter_kws={'alpha':0.4})
plt.title('Votes vs Rating Correlation')
plt.xlabel('Number of Votes')
plt.ylabel('Rating')
plt.grid()
plt.tight_layout()
plt.show()
```

### Conclusion

- After removing **1,094 restaurants with zero votes**, the dataset becomes more reliable for correlation analysis. The **Pearson correlation** between votes and ratings is **0.41**, indicating a **moderate positive relationship**.

- The **scatter plot** supports this finding: it shows a clear **upward trend**, where restaurants with more votes tend to have **better ratings**. Most restaurants are clustered in the **low-vote range**, but the few **high-vote restaurants** often have ratings above **4.5**, reinforcing the link between **popularity (votes)** and **customer satisfaction (ratings)**.

## Task 3

### Task: Price Range vs. Online Delivery and Table Booking
- Analyze if there is a relationship between the price range and the availability of online delivery and table booking.
- Determine if higher-priced restaurants are more likely to offer these services.

```python

# 1. Clean data
df_clean = df[["Price range", "Has Online delivery", "Has Table booking"]].dropna()

# 2. Replace price ranges with category labels
price_labels = {
    1: "Cheap",
    2: "Moderate",
    3: "Expensive",
    4: "Very Expensive"
}
df_clean["Price Label"] = df_clean["Price range"].map(price_labels)

 # 3. Heatmap

# pivot table for both features using label
pivot_online = df_clean.pivot_table(index='Price Label', columns='Has Online delivery', aggfunc='size', fill_value=0)
pivot_booking = df_clean.pivot_table(index='Price Label', columns='Has Table booking', aggfunc='size', fill_value=0)

# Combine both into one DataFrame
heatmap_data = pd.concat([pivot_online, pivot_booking], axis=1)
heatmap_data.columns = ['OD_No', 'OD_Yes', 'TB_No', 'TB_Yes']  # rename for clarity - OD= Online delivery, TB= Table booking  

# prepare correct order
heatmap_data = heatmap_data.reindex(["Cheap", "Moderate", "Expensive", "Very Expensive"])

# Plot heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(heatmap_data, annot=True, fmt="d", cmap="Blues")
plt.title("Service Availability by Price Category (Online Delivery & Table Booking)")
plt.xlabel("Service Type and Availability")
plt.ylabel("Price Category of Restaurants")
plt.show()
```

### Summary 
- As the price category increases, restaurants are less likely to offer online delivery, but more likely to offer table booking.
- Cheap restaurants offer almost no services, whereas Very Expensive restaurants prioritize dine-in reservations over delivery.

```python
# Bar Chart: Online Delivery Availability by Price Category

online_counts = df_clean.groupby(["Price Label", "Has Online delivery"]).size().unstack(fill_value=0)
online_counts = online_counts.reindex(["Cheap", "Moderate", "Expensive", "Very Expensive"])

online_counts.plot(kind='bar', figsize=(8, 5), color=["salmon", "lightgreen"])
plt.title("Online Delivery Availability by Price Category")
plt.xlabel("Price Category")
plt.ylabel("Number of Restaurants")
plt.legend(title="Online Delivery")
plt.xticks(rotation=25)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

### Summary:
- Online delivery is most available in **moderate-priced** restaurants.  
- **Cheap** restaurants rarely provide this service, and **very expensive** ones tend to avoid it.  
- This suggests that mid-range restaurants focus more on **reach and convenience**, while high-end ones prefer dine-in experience.


```python
# Bar Chart: Table Booking

booking_counts = df_clean.groupby(["Price Label", "Has Table booking"]).size().unstack(fill_value=0)
booking_counts = booking_counts.reindex(["Cheap", "Moderate", "Expensive", "Very Expensive"])

booking_counts.plot(kind='bar', figsize=(8, 5), color=["orange", "skyblue"])
plt.title("Table Booking Availability by Price Category")
plt.xlabel("Price Category")
plt.ylabel("Number of Restaurants")
plt.legend(title="Table Booking")
plt.xticks(rotation=25)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

### Summary :
- Table booking is **almost unavailable** in cheap and moderate-priced restaurants.  
- It becomes **significantly available** in expensive and very expensive categories.  
- This shows that **higher-end restaurants prioritize reservations** to manage dine-in experience better.


### Conclusion

The analysis clearly shows a **relationship between price range and service availability**:

-  **Cheap and Moderate restaurants** mostly **do not offer** online delivery or table booking. Their focus seems to be on walk-in or takeout business models.
-  **Expensive restaurants** show a balanced trend, with a noticeable **increase in table booking availability** but a drop in online delivery.
-  **Very Expensive restaurants** rarely offer online delivery, but **almost half support table booking**, reflecting a preference for premium dine-in experiences.

 **Insight:** As price increases, restaurants tend to shift from online convenience to **exclusive in-person service**, especially via **table reservations**.
